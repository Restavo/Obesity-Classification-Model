```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import warnings
import pickle
```

```python
df = pd.read_csv("ObesityDataset1.csv")
df
```

# Preprocessing Dataset / EDA

```python
# Cek informasi dataset
df.info()
```

```python
# Cek jumlah missing value untuk setiap fitur / kolom
df.isna().sum()
```

```python
# Visualisasi menggunakan boxplot untuk mengetahui apakah terdapat Outlier

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['TUE'])
plt.title('Box Plot "TUE"')
plt.xlabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['CAEC'])
plt.title('Box Plot "CAEC"')
plt.xlabel('Frequency')
plt.show()
```

#### Tidak terdapat outlier pada fitur "TUE" dan terdapat outlier pada fitur "CAEC". Maka untuk handle missing value pada fitur "TUE" akan di fill dengan menggunakan Mean dan pada fitur "CAEC" akan di fill dengaan menggunakan Modus karena fitur ini bertipe Kategorik Ordinal

```python
# Imputasi Missing Value
df['TUE'].fillna(df['TUE'].median(), inplace=True)
df['CAEC'].fillna(df['CAEC'].mode()[0], inplace=True)
```

```python
# Cek nilai Unique pada fitur "Age"
df['Age'].unique()
```

```python
# Mengubah tipe data fitur "Age" yang terdapat "years"nya
if df['Age'].dtype == 'object':
    df['Age'] = df['Age'].str.replace(' years', '').astype(float)
df['Age'] = df['Age'].astype(int)
```

```python
# Split fitur (X) dan target (y)
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']
```

```python
# Menggunakan LabelEncoder untuk mengubah target menjadi numeric
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Memisahkan data menjadi Data Training dan Data Testting
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
# Mendefinisikan fitur berdasarkan tipe datanya
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
onehot_features = ['Gender', 'family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC', 'MTRANS']
ordinal_features = ['CAEC', 'CALC']
```

```python
# Menentukan urutan kategori untuk OrdinalEncoder
caec_order = ['no', 'Sometimes', 'Frequently', 'Always']
calc_order = ['no', 'Sometimes', 'Frequently', 'Always']
```

```python
# Filter urutan berdasarkan nilai unik yang ada di data untuk menghindari error
unique_calc_values = df['CALC'].unique()
filtered_calc_order = [order for order in calc_order if order in unique_calc_values]
```

```python
# Membuat preprocessor dengan ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat_onehot', OneHotEncoder(handle_unknown='ignore'), onehot_features),
        ('cat_ordinal', OrdinalEncoder(categories=[caec_order, filtered_calc_order]), ordinal_features)
    ],
    remainder='passthrough'
)
```

# Modeling

### Model 1) Logistic Regression

```python
# Membuat pipeline untuk MOdel Logistic Regression
pipeline_lr = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
])
```

```python
# Training Model
pipeline_lr.fit(X_train, y_train)
```

```python
# Membuat Prediksi
y_pred_lr = pipeline_lr.predict(X_test)
```

```python
print("Classification Report for Logistic Regression Model")
print(classification_report(y_test, y_pred_lr, target_names=le.classes_))
```

### Model 2) Random Forest

```python
# Membuat pipeline untuk Model Random Forest
pipeline_rf = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
# Training model
pipeline_rf.fit(X_train, y_train)
```

```python
# Membuat prediksi
y_pred_rf = pipeline_rf.predict(X_test)
```

```python
print("Classification Report for Random Forest Model")
print(classification_report(y_test, y_pred_rf, target_names=le.classes_))
```

#### Pada kasus ini, model yang terbaik adalah "Random Forest". Karena jika dilihat dari "accuracy", "macro avg", dan "weighted avg", model "Random Forest" sedikit lebih unggul dibandingkan model "Logistic Regression".

# Save Best Model

```python
model_filename = 'obesity_classification_pipeline.pkl'
encoder_filename = 'obesity_label_encoder.pkl'
```

```python
# Save pipeline dari model RF
with open(model_filename, 'wb') as file:
    pickle.dump(pipeline_rf, file)
```

```python
# Save label encoder
with open(encoder_filename, 'wb') as file:
    pickle.dump(le, file)
```
